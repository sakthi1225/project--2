###**DATASET INFORMATION**

---
####**About DataSet:**
This dataset consists of 10,001 rows and 15 columns, capturing detailed insights into customer transactions and purchasing behavior. It is designed to help analyze customer engagement, sales performance, and retention strategies by exploring various customer-centric metrics. The data spans different regions and includes time-based elements such as launch and peak sales dates, offering opportunities for seasonal and trend analysis.

####**Attribute Information**
1. **Customer_Id**   - Unique identifier for each customer.
2. **Purchase_Id**   - Unique identifier for each purchased product.
3. **Transaction_Id** - Unique identifier for each transaction.
4. **Purchase_Frequency** - Total number of purchases made by the customer.
5. **Average_Order_Value**	- Average value of the customer's orders in monetary terms.
6. **Most_Frequent_Category** -	Product category most often purchased by the customer (e.g., Electronics,clothing,sports).
7. **Time_Between_Purchases** - Average time gap (in days) between two purchases.
8. **Region**	- Geographic region of the customer (e.g., Asia, Europe,North America,South America).
9. **Churn_Probability**	- Probability (0 to 1) that the customer will stop purchasing.
10. **Lifetime_Value** - Estimated long-term monetary value of the customer.
11. **Launch_Date** - Date when the customer first interacted with the platform/product.
12. **Peak_Sales_Date** - Date when the customer made their highest value purchase.
13. **Season** - Season associated with peak purchasing activity (e.g., Spring, Winter,Summer,Fall).
14. **Preferred_Purchase_Times** - Time of day the customer prefers to shop (e.g., Morning,Afternoon, Evening).
15. **Retention_Strategy** - Strategy used to retain the customer (e.g., Loyalty Program, Discount,Email campaign).

✉**ENCODING**

 **one hot encoding**

```python
from sklearn.preprocessing import LabelEncoder#import labelencoder
# Create a new column: if churn probability > 0.5, mark as 'Churned', else 'Not Churned'
df['Churn_Pro'] = df['Churn_Probability'].apply(lambda x: 'Churned' if x > 0.5 else 'Not Churned')
# Create a LabelEncoder object
encoder=LabelEncoder()
# Convert 'Churn_Pro' text labels into numbers (like 0 and 1)
df['Churn_Encoded']=encoder.fit_transform(df['Churn_Pro'])
print(df)
```

**LabelEncoding( Encoding Retention_Strategy column)**

```python
le = LabelEncoder()
# Fit and transform
df['Retention_Strategy_Encoded'] = le.fit_transform(df['Retention_Strategy'])
# Display label mapping
label_map = dict(zip(le.classes_, le.transform(le.classes_)))
print("Label Mapping:", label_map)
# Show updated DataFrame
print(df.head())
```

```python
print(df.columns)
```

```python
df = df.drop(['Churn_Pro'], axis=1)# Dropping intermediate churn label column
print(df)
```

**ONE_HOT_ENCODING( One-hot encoding Season column)**

```python
oh = OneHotEncoder(sparse_output=False)
season_reshaped = df['Season'].values.reshape(-1, 1)# Reshape the 'Season' column for encoding
encoded_season = oh.fit_transform(season_reshaped)# Apply one-hot encoding to the reshaped 'Season' column
encoded_df = pd.DataFrame(encoded_season, columns=oh.categories_[0], dtype=int)# Create a new DataFrame with encoded columns and proper labels
df = pd.concat([df.drop('Season', axis=1), encoded_df], axis=1) #Remove the original 'Season' column and add encoded columns
print("After Encoding:\n", df.head(20))  # Display the first 20 rows of the updated DataFrame
```

#**MACHINE LEARNING MODEL**

🔎**FEATURE SELECTION**

```python
#import necessary libraries
from sklearn.model_selection import train_test_split
from sklearn.feature_selection import SelectKBest, f_classif
from sklearn.preprocessing import StandardScaler
# Define features (X) and target (y)
X = df.drop(columns=['Churn_Encoded', 'Customer_ID', 'Product_ID', 'Transaction_ID', 'Launch_Date','Region','Peak_Sales_Date','Preferred_Purchase_Times'])
y = df['Churn_Encoded']
# Initialize the SelectKBest feature selector
selector = SelectKBest(score_func=f_classif, k=20)
```

✅**using SELECTKBEST**

```python
X_new = selector.fit_transform(X, y)#select top k features
selected_features = X.columns[selector.get_support()]  # Changed selecter to selector
features_score = pd.DataFrame({'features': X.columns, 'selected': selector.get_support()})  # Changed selecter to selector
print(selected_features)
print(features_score)
```

```python
#Split dataset into training and testing (70% - 30%)
X_train, X_test, y_train, y_test = train_test_split(X_new,y,test_size=0.3, random_state=42) # 70% train, 30% test
```

```python
Ss=StandardScaler()
X_test_scaled=Ss.fit_transform(X_test)
X_train_scaled=Ss.fit_transform(X_train)
# Now, X_train_scaled and X_test_scaled are ready for model training and evaluation
print("Scaled Training Data:")
print(X_train_scaled[:5])  # Displaying first 5 rows of the scaled training data
print("Scaled Testing Data:")
print(X_test_scaled[:5])  # Displaying first 5 rows of the scaled testing data
```

**MODEL BUILDING**

```python
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.svm import SVC
from sklearn.ensemble import RandomForestClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report
```

```python
#Initialize classifiers in a dictionary
classifiers={
    'Logistic Regression':LogisticRegression(),
    'Decision Tree':DecisionTreeClassifier(),
    'SVM':SVC(),
    'RFC':RandomForestClassifier(),
    'KNN':KNeighborsClassifier()

}
```

```python
for name,clf in classifiers.items():
  clf.fit(X_train_scaled,y_train)#Train the model
  y_pred=clf.predict(X_test_scaled) #predict on the test data
  accuracy=accuracy_score(y_test,y_pred)#evaluate accuray
  print(f"Accuracy of {name}:{accuracy}")
  print("/n Confusion matrix:")
  print(confusion_matrix(y_test,y_pred))#evaluate confusion matrix
  print("/n Classification report")
  print(classification_report(y_test,y_pred))#classification report
```

🏆**Identifying the Best Classifier**

```python
best_accuracy=0 #Initialize variables to store the best model and its accuracy
best_classifier=None
if accuracy>best_accuracy:
  best_accuracy=accuracy
  best_classifier=name
print(f"\nMost efficient Classifier:{best_classifier}with accuracy{best_accuracy}") #Display the best-performing classifier and its accuracy
```
